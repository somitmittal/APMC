```python
import datetime
months={}
for i in range(1,13):
    months[datetime.date(2008, i, 1).strftime('%B')]=i
months
```

```python
df2=pd.read_csv(dpath2)
df2['commodity']=df2['commodity'].str.upper()
crops=dict(zip(df2['commodity'],df2['Type']))
crops
```

```python
df_ahmednagar=df_temp1.loc[df_temp1['APMC']=="Ahmednagar"]
df_ahmednagar
df_ahmednagar['Month'].replace(months,inplace=True)
df_ahmednagar.head()
```

```python
df_ahmednagar['Commodity']=df_ahmednagar['Commodity'].str.upper()
df_ahmednagar["Commodity"].unique()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(name_commodity):
    if name_commodity in crops:
        type_of_crop=crops[name_commodity]
        df_a_bajri=df_ahmednagar.loc[df_ahmednagar['Commodity']==name_commodity]
        df_a_bajri.sort_values(['Year','Month'],inplace=True)
        df_a_bajri.plot(x=['Year','Month'],y=['min_price','modal_price','max_price'],kind='bar',title='Modal Prices for ' + name_commodity+' at Ahmednagar APMC in Ahamadnagar District Maharashtra Type of crop= '+type_of_crop)
        return df_a_bajri
    else:
        print "Crop not found"
```

```python
df_big=f('MAIZE')
```

```python
from sklearn import linear_model
def plot_data(df_plot):
    df=df_plot[['Year','Month','modal_price']]
    x_data=np.array(df[['Year','Month']])
    y_data=np.array(list(df['modal_price']))
    x_test=np.array([(2016,11),(2016,12),(2017,1)])
    reg=linear_model.LinearRegression()
    reg.fit(x_data,y_data)
    pred=reg.predict(x_test)
    x_test_df=pd.DataFrame(list(x_test),columns=['Year','Month'])
    x_test_df['modal_price']=pred
    df=df.append(x_test_df)
    df.plot(x=['Year','Month'],y=['modal_price'],kind='bar',title='Training Modal Price at Ahmednagar APMC')
```

```python
plot_data(df_big)
```
